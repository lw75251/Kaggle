### Missing Data 

#### Combine train and test to see which columns to has the most missing

```python
y_train = train['SalePrice']
test_id = test['Id']
all_data = pd.concat([train, test], axis=0, sort=False)
all_data = all_data.drop(['Id', 'SalePrice'], axis=1)
```

```python
Total = all_data.isnull().sum().sort_values(ascending=False)
percent = (all_data.isnull().sum() / all_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([Total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(25)
```

The features that have the most missing values are not the ones that have the strongest positive correlations. So, we can just delete them as we are not using it.

```python
all_data.drop((missing_data[missing_data['Total'] > 5]).index, axis=1, inplace=True)
print(all_data.isnull().sum().max())
```

Let's show the features and the number of Missing values

```python
total = all_data.isnull().sum().sort_values(ascending=False)
total.head(19)
```

### Filling in Missing Data

```python
# Fill Numeric Data with 0s
numeric_missed = ['BsmtFinSF1',
                  'BsmtFinSF2',
                  'BsmtUnfSF',
                  'TotalBsmtSF',
                  'BsmtFullBath',
                  'BsmtHalfBath',
                  'GarageArea',
                  'GarageCars']

for feature in numeric_missed:
    all_data[feature] = all_data[feature].fillna(0)
```

```python
#Fill Categorical Data with respective Mode
categorical_missed = ['Exterior1st',
                  'Exterior2nd',
                  'SaleType',
                  'MSZoning',
                   'Electrical',
                     'KitchenQual']

for feature in categorical_missed:
    all_data[feature] = all_data[feature].fillna(all_data[feature].mode()[0])
```

```python
#Fill all other data with respective most common value value
all_data['Functional'] = all_data['Functional'].fillna('Typ')
```

```python
all_data.drop(['Utilities'], axis=1, inplace=True)
```

And now, we have 0 missing data:

```python
all_data.isnull().sum().max() #just checking that there's no missing data missing...
```

## 4. Feature Engineering

#### Fix The Skewness in the other features


```python
numeric_feats = all_data.dtypes[all_data.dtypes != 'object'].index
skewed_feats = all_data[numeric_feats].apply(lambda x: skew(x)).sort_values(ascending=False)
high_skew = skewed_feats[abs(skewed_feats) > 0.5]
high_skew
```

```python
for feature in high_skew.index:
    all_data[feature] = np.log1p(all_data[feature])
```

### Feature Engineering

```python
# Combines all SquareFootage attributes into one
all_data['TotalSF'] = all_data['TotalBsmtSF'] + all_data['1stFlrSF'] + all_data['2ndFlrSF']
```

## Model Preparation

### One-Hot Encoding

```python
all_data = pd.get_dummies(all_data)
all_data.head()
```

```python
x_train =all_data[:len(y_train)]
x_test = all_data[len(y_train):]
```
